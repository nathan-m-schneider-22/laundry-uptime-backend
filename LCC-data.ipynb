```python
import requests
import matplotlib.pyplot as plt
import datetime
```

```python
import requests

x = requests.get('https://laundry-uptime.herokuapp.com/data')
data = x.json()
labelcount = min(20,len(data))
indices = []
for i in range(labelcount):
    indices.append(int(i*len(data)/labelcount))

data.sort(key = lambda i: i['time'])
times = [d["time"] for d in data]
dates = [datetime.datetime.fromtimestamp(t/1000).strftime("%m/%d, %H:%M") for t in times]
counts = [d["count"] for d in data]

plt.plot(times,counts)
plt.xticks([times[t] for t in indices],labels = [dates[t] for t in indices],rotation='vertical')
plt.title("LCC Climbers")
print("Total Entries: " ,len(data))
# for i in range(len(data)): print(i,data[i])
```
